```python
from functools import partial
from random import choice, randrange
from string import ascii_uppercase as letters
from string import digits

from faker import Faker

letters = "АВЕКМНОРСТУХ"
```

```python
fake = Faker("ru_RU")
n = 100
```

```python
def oleg(s):
    match s:
        case "d":
            return choice(digits)
        case "s":
            return choice(letters)
        case _:
            return s

def generate(pattern: str) -> str:
    return "".join(map(oleg, pattern))


generate("+7 ddd ddd-dd-dd")
```

```python
def joins(*items: list) -> str:
    return f"({', '.join(map("'{}'".format, items))})"

def pprint(items):
    print(*(joins(i, *items()) for i in range(n)),
          sep=",\n", end=";\n")

file = open("script_data.txt", 'w', encoding="utf-8")
frint = partial(print, file=file)
def write(who, items):
    pprint(items)
    frint(f"INSERT INTO {who} VALUES")
    frint(*(joins(i, *items()) for i in range(n)), 
          sep=",\n", end=";\n")
    frint()
    file.flush()
```

```python
write("Client", lambda: (
    fake.name(),
    generate("dd dd dddddd"),
    fake.date_of_birth(minimum_age=22),
    generate("+7 ddd ddd-dd-dd")
))
```

```python
write("Car", lambda: (
    fake.text(max_nb_chars=20),
    randrange(1990, 2024),
    fake.color_name(),
    generate("s ddd ss ddd"),
))
```

```python
write("Service", lambda: (
    fake.company(),
    fake.address(),
    fake.bban()
))
```

```python
write("Rent", lambda: (
    randrange(n),
    randrange(n),
    *sorted((
        fake.date_time_this_month(),
        fake.date_time_this_month(),
    )),
    randrange(100, 10000)
))
```

```python
write("Maintenance", lambda: (
    randrange(n),
    randrange(n),
    *sorted((
        fake.date_time_this_month(),
        fake.date_time_this_month(),
    )),
    randrange(100, 10000)
))
```
