```python
from random import randrange

from faker import Faker
```

```python
fake = Faker("ru_RU")
n = 10
```

```python
def joins(*items: list) -> str:
    return f"({', '.join(map("'{}'".format, items))})"

def pprint(items):
    print(*(joins(i, *items()) for i in range(n)), sep=",\n")
```

```python
pprint(lambda: (
    fake.name(),
    "{} {} {}".format(randrange(10, 100),
                        randrange(10, 100),
                        randrange(100_000, 1_000_000)),
    fake.date_this_century(),
    fake.phone_number()
))
```

```python
pprint(lambda: (
    fake.text(max_nb_chars=20),
    randrange(1990, 2024),
    fake.color_name(),
    fake.license_plate()
))
```

```python
pprint(lambda: (
    fake.company(),
    fake.address(),
    fake.bban()
))
```

```python
pprint(lambda: (
    randrange(10),
    randrange(10),
    *sorted((
        fake.date_time_this_month(),
        fake.date_time_this_month(),
    )),
    randrange(100, 10000)
))
```
